# Hypothesis Testing:

## 1. Comparing Means: Independent Samples t-test

### 1- Sample T test


H0: There is no difference in the average bike rentals between weekdays and weekends.


Ha: There is a significant difference in the average bike rentals between weekdays and weekends.

```python
from scipy.stats import ttest_ind

# Example: Independent t-test for two independent samples
stat, p_value = ttest_ind(df[df['weekday'] == 0]['cnt'], df[df['weekday'] == 1]['cnt'])

# Check the results
print(f'Test Statistic: {stat}')
print(f'P-value: {p_value}')

# Make a decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

## 2. Proportions: Z-test for Proportions Hypothesis:

Null Hypothesis (H0): The proportion of holidays with bike rentals is equal to the proportion of non-holidays with bike rentals.

Alternative Hypothesis (Ha): There is a significant difference in the proportions of bike rentals between holidays and non-holidays.

```python
from scipy.stats import ttest_1samp

# Example: One-sample t-test for the mean bike rentals on holidays
expected_value = 100  # Replace with the value you want to test against
stat, p_value = ttest_1samp(df[df['holiday'] == 1]['cnt'], expected_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

```

## 3. ANOVA: Analysis of Variance Hypothesis:

Null Hypothesis (H0): The mean bike rentals are equal across all seasons.
Alternative Hypothesis

(Ha): There is a significant difference in the mean bike rentals across seasons.

```python
from scipy.stats import f_oneway

# Example: One-way ANOVA for multiple groups (seasons)
f_stat, p_value = f_oneway(df[df['season'] == 1]['cnt'], df[df['season'] == 2]['cnt'],
                            df[df['season'] == 3]['cnt'], df[df['season'] == 4]['cnt'])

# Decision
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
